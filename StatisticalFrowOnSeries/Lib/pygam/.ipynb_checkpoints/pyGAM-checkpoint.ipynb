# Data Examples

```python
from pygam import GAM
%matplotlib inline
```

```python
x = np.linspace(-5,5,5000)[:,None]
obs = -.5*x**2 + 5 + 3*np.random.randn(len(x),1)
plt.plot(x, obs)
```

```python
gam = GAM(spline_order=1)
gam.fit(x, obs)
```

```python
gam.summary()
```

```python
plt.plot(x, obs)
plt.plot(x, gam.predict(x))
```

```python
a, b = (gam.prediction_intervals(x, width=.95).T)
((obs.ravel() > a) * (obs.ravel() < b)).mean()
```

```python
plt.plot(x, obs)
plt.plot(x, gam.prediction_intervals(x), color='g')
```

```python
plt.plot(x, obs)
plt.plot(x, gam.confidence_intervals(x), color='g')
```

```python
gam._statistics['pseudo_r2']
```

# Numerical Meth

```python
x = np.linspace(-5,5,5000)[:,None]

log_odds = -5*x**2 + 5# quadratic
# log_odds = 10*x#  -5.77104596# linear
p = 1/(1+np.exp(-log_odds)).squeeze()

obs = (np.random.rand(len(x)) < p).astype(np.int)
# obs[len(x)/2:] = obs[:len(x)/2][::-1]

plt.plot(x, p)
plt.scatter(x, obs)
```

```python
from pygam import LogisticGAM
```

```python
lgam = LogisticGAM(lam=1, n_iter=50, n_knots=50, tol=1e-5, spline_order=1, penalty_matrix='auto')
lgam.fit(x, obs)
```

```python
# plt.plot(p)
plt.scatter(x, obs)
plt.plot(x, lgam.predict_proba(x))
plt.plot(x, p)
```

```python
plt.plot(lgam.logs['deviance'])
```

```python
plt.plot(lgam.logs['diffs'])
```

```python
plt.plot(lgam.logs['accuracy'])
```

```python
plt.plot(lgam.predict_proba(x))
plt.plot(lgam.confidence_intervals(x, .99), color='g')
```

```python
plt.plot(lgam.prediction_intervals(x, .95), color='g')
```

# GCV Loop

```python
scores = []
lams = np.logspace(-2,1, 21)
for lam in lams:
    lgam = LogisticGAM(lam=lam, n_knots=25)
    lgam.fit(x, obs)
    scores.append(lgam._statistics['UBRE'])

plt.plot(lams, scores)
best = np.argmin(scores)
plt.scatter(lams[best], scores[best], facecolor='None')
plt.title('Best Lam: {}'.format(lams[best]))
```

```python
lgam = LogisticGAM(lam=lams[best])
lgam.fit(x, obs)
```

```python
lgam._statistics['edof']
```

```python
plt.plot(x, lgam.confidence_intervals(x, width=0.95), '--', c='b')
plt.plot(x, lgam.predict_proba(x))
```

```python
plt.plot(x, lgam.prediction_intervals(x, width=0.95), '--', c='b')
plt.plot(x, lgam.predict_proba(x))
```

```python
plt.fill_between(x.squeeze(), *lgam.prediction_intervals(x, width=0.95).T, alpha=0.25)
plt.plot(x, lgam.predict_proba(x))
```

```python
plt.plot(x, lgam.predict_proba(x))
plt.scatter(x, obs)
```

```python
plt.plot(lgam._b)
```

```python
plt.plot(lgam.partial_dependence(np.sort(x, axis=0))[0])
```

```python
lgam.accuracy(x, obs)
```

# Multiple Features

```python
n = 10000
x = np.random.rand(n, 5) * 10 - 5
cat = np.random.randint(0,4, n)
x = np.c_[x, cat]

log_odds = (-0.5*x[:,0]**2) + 5 +(-0.5*x[:,1]**2) + np.mod(x[:,-1], 2)*-30
p = 1/(1+np.exp(-log_odds)).squeeze()

obs = (np.random.rand(len(x)) < p).astype(np.int)
```

```python
# lgam = GAM(distribution='binomial', link='logit', lam=0.6, n_iter=200, n_knots=20, spline_order=4, penalty_matrix='auto', tol=1e-4)
lgam = LogisticGAM(lam=0.6, n_iter=100, n_knots=20, spline_order=4, penalty_matrix='auto', tol=1e-4)
lgam.fit(x, obs)
```

```python
plt.plot(lgam.partial_dependence(np.sort(x, axis=0))[0]);
```

```python
plt.plot(lgam.partial_dependence(np.sort(x, axis=0))[0][:,-1]);
plt.plot(lgam.partial_dependence(np.sort(x, axis=0), quantiles=np.linspace(.1, .9, 4))[1][-1]);
```
